```python
#Removing columns that has more than 30%missing values.b
alldata=alldata.drop(columns={'respondent.4','respondent.3','respondent.2','respondent.1','parties.2'},axis=1)
alldata=alldata.drop(columns={'issue.9','issue.8','issue.7','issue.6','issue.5','issue.4','issue.3'},axis=1)
alldata=alldata.drop(columns={'issue.2','issue.1','issue.0','issue.26','issue.25','issue.24','issue.23'},axis=1)
alldata=alldata.drop(columns={'issue.22','issue.21','issue.20','issue.19','issue.18','issue.17','issue.16'},axis=1)
alldata=alldata.drop(columns={'issue.15','issue.14','issue.13','issue.12','issue.11','issue.10'},axis=1)
alldata=alldata.drop(columns={'introductiondate','decisiondate'},axis=1)
```

```python
for col in alldata.columns:
    print(col    ,'---------Unique Values--->',alldata[col].nunique())
```

```python
#Removing column that has only unique numerical value
for col in alldata.columns:
    if len(alldata[col].unique()) ==1:
        alldata.drop(col,inplace=True,axis=1)
```

```python
for col in alldata.columns:
    if alldata[col].dtype=='object':
        alldata[col]=alldata[col].str.lower()
```

```python
is_similar_col=['country.alpha2','country.name','parties.1','respondent.0']
print(alldata[is_similar_col].head())
#print(alldata[is_similar_col].tail(10))
```

```python
#The above 4 columns are same so removing three
alldata=alldata.drop(columns={'country.alpha2','country.name','parties.1'},axis=1)
```

```python
#columns docname and parites.0 is no useful 
alldata=alldata.drop(columns={'docname','parties.0'},axis=1)

#columns judgement date and kp date are same hence removing one
alldata=alldata.drop(columns={'kpdate'},axis=1)
```

```python
#removing "001-" in itemid column
alldata['itemid']=alldata['itemid'].apply(lambda x: str(x)[4:10])
```

```python
#keep only values before decimal in rank
alldata['rank']=alldata['rank'].apply(lambda x: str(x).split('.')[0])
```

```python
#since i have removed decisiondate and introduction date so i will remove judgement date too
#because i think we can make no new extra features
#Will do feature engineering in another notebook

#Extracting Datetime features 
#extracting brought no value
alldata=alldata.drop('judgementdate',axis=1)
```

```python
#Encoding doctypebranch,originatingbody_name,respondent.0,separateopinion
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
alldata['doctypebranch']=le.fit_transform(alldata['doctypebranch'])
alldata['originatingbody_name']=le.fit_transform(alldata['originatingbody_name'])
alldata['respondent.0']=le.fit_transform(alldata['respondent.0'])
alldata['separateopinion']=le.fit_transform(alldata['separateopinion'])
```

```python
#only ecli column left for cleaning
#i haven't understood how to deal with ecli column so i will just remove it too  😁


alldata=alldata.drop('ecli',axis=1)
```
